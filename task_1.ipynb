```python
def hypothesis(X, theta):
    return np.dot(X, theta)
```

```python
def cost_function(X, y, theta):
    return np.square(hypothesis(X, theta) - y).sum() / (2 * X.size)
```

```python
def gradient_step(X, y, theta, learning_rate):
    m = len(y)
    predictions = np.dot(X, theta)
    errors = predictions - y
    gradient = (1 / m) * np.dot(X.T, errors)
    theta = theta - learning_rate * gradient
    return theta
```

```python
def gradient(X, y, learning_rate, iterations, eps):
    ones = np.ones((X.shape[0], 1))
    X = np.hstack((ones,X))

    theta = np.zeros(X.shape[1])

    cost = cost_function(X, y, theta)
    cost_history = [cost]
    for i in range(iterations):
        theta = gradient_step(X, y, theta, learning_rate)
        cost = cost_function(X, y, theta)

        if abs(cost - cost_history[-1]) < eps:
            cost_history.append(cost)
            break

        cost_history.append(cost)

    return theta, cost_history
```

```python
norm = df.copy()
columns = ['price', 'area', 'bedrooms', 'bathrooms']
for column in columns:
    norm[column] = (df[column] - df[column].mean()) / df[column].std()

X = norm[['area', 'bedrooms', 'bathrooms']].values
y = norm['price'].values

learning_rate = 0.001
iterations = 10000
eps = 1e-12

theta, cost_history = gradient(X, y, learning_rate, iterations, eps)

print("Optimal parameters (theta):", theta.round(7))
```

```python
theta_analytical = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("Optimal parameters (theta):", theta_analytical.round(7))
```

```python
# Створення та навчання моделі лінійної регресії з scikit-learn
model = LinearRegression()
model.fit(X, y)

# Прогнозування цін на будинки
predicted_prices_sklearn = model.predict(X)

# Параметри, знайдені за допомогою sklearn
theta_sklearn = np.hstack((model.intercept_, model.coef_))

print("Optimal parameters (theta):", theta_sklearn.round(7))
```
